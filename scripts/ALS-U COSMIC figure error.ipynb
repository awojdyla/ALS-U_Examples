#### Change the scale of a randomly generated mirror (1nm-rms mfe)

```python
def scale_mirror(filename_in, scale_nm, filename_out):
    fid = open(filename_in)
    a = fid.readline()
    b = fid.readline()

    array_size = numpy.fromstring(a, sep=' ',dtype='int32')
    x_m = numpy.fromstring(b, sep=' ')
    y_m = numpy.zeros(array_size[0])
    Z_m = numpy.zeros((array_size[0],array_size[1]))
    for il in numpy.arange(array_size[0]):
        c = fid.readline()
        C = numpy.fromstring(c, sep=' ')
        y_m[il]   = C[0]
        Z_m[il,:] = C[1:]*scale_nm
    Shadow.ShadowTools.write_shadow_surface(numpy.transpose(Z_m),y_m, x_m, outFile=filename_out)
```

#### OASYS generated Shadow code

```python
def propagate_cosmic_shadow(x_source_um, y_source_um, mfe_scale_nm, x_div_rad = 12e-6, y_div_rad = 12e-6):

    
    filename_in  = b'/Users/awojdyla/Oasys/ALS-U_Examples/workspaces/cosmic-transition/mirror100.dat'
    filename_out = '/Users/awojdyla/Oasys/ALS-U_Examples/workspaces/cosmic-transition/mirror101s.dat'
    scale_mirror(filename_in, mfe_scale_nm, filename_out)
    filename_in  = b'/Users/awojdyla/Oasys/ALS-U_Examples/workspaces/cosmic-transition/mirror200.dat'
    filename_out = '/Users/awojdyla/Oasys/ALS-U_Examples/workspaces/cosmic-transition/mirror102s.dat'
    scale_mirror(filename_in, mfe_scale_nm, filename_out)
    filename_in  = b'/Users/awojdyla/Oasys/ALS-U_Examples/workspaces/cosmic-transition/mirror300.dat'
    filename_out = '/Users/awojdyla/Oasys/ALS-U_Examples/workspaces/cosmic-transition/mirror103s.dat'
    scale_mirror(filename_in, mfe_scale_nm, filename_out)

    iwrite = 0

    #
    # initialize shadow3 source (oe0) and beam
    #
    beam = Shadow.Beam()
    oe0 = Shadow.Source()
    oe1 = Shadow.OE()
    oe2 = Shadow.OE()
    oe3 = Shadow.OE()
    oe4 = Shadow.OE()
    oe5 = Shadow.OE()

    #
    # Define variables. See meaning of variables in: 
    #  https://raw.githubusercontent.com/srio/shadow3/master/docs/source.nml 
    #  https://raw.githubusercontent.com/srio/shadow3/master/docs/oe.nml
    #

    oe0.FDISTR = 3
    oe0.FSOURCE_DEPTH = 2
    oe0.F_PHOT = 0
    oe0.HDIV1 = 0.0
    oe0.HDIV2 = 0.0
    oe0.IDO_VX = 0
    oe0.IDO_VZ = 0
    oe0.IDO_X_S = 0
    oe0.IDO_Y_S = 0
    oe0.IDO_Z_S = 0
    oe0.ISTAR1 = 0
    oe0.NPOINT = 100000
    oe0.PH1 = 260.0
    oe0.SIGDIX = 3.5e-05
    oe0.SIGDIZ = 3.5e-05
    oe0.SIGMAX = x_source_um*1e-6
    oe0.SIGMAZ = y_source_um*1e-6
    oe0.VDIV1 = 0.0
    oe0.VDIV2 = 0.0
    oe0.WYSOU = 1.0

    oe1.ALPHA = 90.0
    oe1.DUMMY = 100.0
    oe1.FHIT_C = 1
    oe1.FILE_RIP = b'/Users/awojdyla/Oasys/ALS-U_Examples/workspaces/cosmic-transition/mirror101s.dat'
    oe1.FMIRR = 3
    oe1.F_EXT = 1
    oe1.F_G_S = 2
    oe1.F_RIPPLE = 1
    oe1.RLEN1 = 0.125
    oe1.RLEN2 = 0.125
    oe1.RWIDX1 = 0.01
    oe1.RWIDX2 = 0.01
    oe1.R_MAJ = 1000000.0
    oe1.R_MIN = 0.654
    oe1.T_IMAGE = 0.0
    oe1.T_INCIDENCE = 88.75
    oe1.T_REFLECTION = 88.75
    oe1.T_SOURCE = 15.0

    oe2.ALPHA = 90.0
    oe2.DUMMY = 100.0
    oe2.T_IMAGE = 0.0
    oe2.T_INCIDENCE = 88.5
    oe2.T_REFLECTION = 88.5
    oe2.T_SOURCE = 2.6

    oe3.ALPHA = 180.0
    oe3.DUMMY = 100.0
    oe3.F_CENTRAL = 1
    oe3.F_GRATING = 1
    oe3.F_MONO = 2
    oe3.PHOT_CENT = 260.0
    oe3.RULING = 100000.0
    oe3.R_LAMBDA = 5000.0
    oe3.T_IMAGE = 0.0
    oe3.T_INCIDENCE = 88.5
    oe3.T_REFLECTION = 86.736
    oe3.T_SOURCE = 0.0

    oe4.ALPHA = 90.0
    oe4.DUMMY = 100.0
    oe4.FHIT_C = 1
    oe4.FILE_RIP = b'/Users/awojdyla/Oasys/ALS-U_Examples/workspaces/cosmic-transition/mirror102s.dat'
    oe4.FMIRR = 3
    oe4.F_EXT = 1
    oe4.F_G_S = 2
    oe4.F_RIPPLE = 1
    oe4.RLEN1 = 0.175
    oe4.RLEN2 = 0.175
    oe4.RWIDX1 = 0.01
    oe4.RWIDX2 = 0.01
    oe4.R_MAJ = 305.6
    oe4.R_MIN = 0.262
    oe4.T_IMAGE = 5.0
    oe4.T_INCIDENCE = 88.5
    oe4.T_REFLECTION = 88.5
    oe4.T_SOURCE = 2.4

    oe5.DUMMY = 100.0
    oe5.FHIT_C = 1
    oe5.FILE_RIP = b'/Users/awojdyla/Oasys/ALS-U_Examples/workspaces/cosmic-transition/mirror103s.dat'
    oe5.FMIRR = 3
    oe5.F_EXT = 1
    oe5.F_G_S = 2
    oe5.F_RIPPLE = 1
    oe5.RLEN1 = 0.175
    oe5.RLEN2 = 0.175
    oe5.RWIDX1 = 0.01
    oe5.RWIDX2 = 0.01
    oe5.R_MAJ = 116.647
    oe5.R_MIN = 0.08
    oe5.T_IMAGE = 2.2
    oe5.T_INCIDENCE = 88.5
    oe5.T_REFLECTION = 88.5
    oe5.T_SOURCE = 5.0



    #Run SHADOW to create the source

    if iwrite:
        oe0.write("start.00")

    beam.genSource(oe0)

    if iwrite:
        oe0.write("end.00")
        beam.write("begin.dat")


    #
    #run optical element 1
    #
    print("    Running optical element: %d"%(1))
    if iwrite:
        oe1.write("start.01")

    beam.traceOE(oe1,1)

    if iwrite:
        oe1.write("end.01")
        beam.write("star.01")


    #
    #run optical element 2
    #
    print("    Running optical element: %d"%(2))
    if iwrite:
        oe2.write("start.02")

    beam.traceOE(oe2,2)

    if iwrite:
        oe2.write("end.02")
        beam.write("star.02")


    #
    #run optical element 3
    #
    print("    Running optical element: %d"%(3))
    if iwrite:
        oe3.write("start.03")

    beam.traceOE(oe3,3)

    if iwrite:
        oe3.write("end.03")
        beam.write("star.03")


    #
    #run optical element 4
    #
    print("    Running optical element: %d"%(4))
    if iwrite:
        oe4.write("start.04")

    beam.traceOE(oe4,4)

    if iwrite:
        oe4.write("end.04")
        beam.write("star.04")


    #
    #run optical element 5
    #
    print("    Running optical element: %d"%(5))
    if iwrite:
        oe5.write("start.05")

    beam.traceOE(oe5,5)

    if iwrite:
        oe5.write("end.05")
        beam.write("star.05")   
    
    return beam
```

```python
%%capture
source_size_um = numpy.arange(11)*25
mfe_scale_nm   = numpy.arange(11)*5
x_spot_size_um = numpy.zeros((source_size_um.size, mfe_scale_nm.size))
y_spot_size_um = numpy.zeros((source_size_um.size, mfe_scale_nm.size))

for i_source in numpy.arange(source_size_um.size):
    for i_scale in numpy.arange(mfe_scale_nm.size):
        beam = propagate_cosmic_shadow(x_source_um=source_size_um[i_source], y_source_um=source_size_um[i_source],
                                       mfe_scale_nm=mfe_scale_nm[i_scale], x_div_rad = 12e-6, y_div_rad = 12e-6)
        #Shadow.ShadowTools.plotxy(beam,1,3,nbins=101,nolost=1,title="Real space")
        x_FWHM_um = beam.get_standard_deviation(1)*2.35e6
        y_FWHM_um = beam.get_standard_deviation(3)*2.35e6
        
        x_spot_size_um[i_source, i_scale] = x_FWHM_um
        y_spot_size_um[i_source, i_scale] = y_FWHM_um
```

```python
import matplotlib.pyplot as plt
plt.cla()
plt.plot(source_size_um/2.35, x_spot_size_um/2.35)
plt.xlabel('source size [um-rms]')
plt.ylabel('y spot size [um-rms]')
plt.title('COSMIC, 0 to 50nm-rms mirror figure error ')
plt.xlim((0,100))
plt.ylim((0,100))
plt.axes().set_aspect('equal')
#for line in plt.gca().get_lines() : line.set_color("black"); i+=1
plt.savefig('COSMIC_spot_size_y',dpi=300,transparent=True)
plt.show()
#plt.plot(y_spot_size_um)
```
